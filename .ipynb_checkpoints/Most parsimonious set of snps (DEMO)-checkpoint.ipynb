# load snps from vcf files

```python
import allel
from os import path
import sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scanpy as sc
from anndata import AnnData


data_dir = '/home/hsher/data'
fname = 'ALL.TGP_HGDP.GRCh37.bi.sorted.fixref.phased.isec.23andMe.5000-1000-0.01_pairwise.vcf.gz'
tabix = '/usr/local/bin/tabix'
info = allel.read_vcf(path.join(data_dir, fname),log=sys.stdout)

```

```python
sample = info['samples'] # sample ID
```

```python
#to genotype array
gt = allel.GenotypeArray(info['calldata/GT'], dtype = 'i1')
gt = gt.to_n_alt(fill = -1)
```

# first type of imputing is just using the most common allele in "whole population"

```python
from sklearn.impute import SimpleImputer
imp_mean = SimpleImputer(missing_values=-1, strategy='most_frequent')
gt_impute = imp_mean.fit_transform(gt)
```

# load metadata

```python
# read 
sample_info = pd.read_pickle(path.join(data_dir, 'sample_info.pickle'))

# load custom colors
from custome_color import *
```

```python
# to use scanpy, everything has to go into AnnData, see https://anndata.readthedocs.io/en/stable/anndata.AnnData.html
adata = AnnData(X = gt.T, obs = sample_info.loc[sample])
adata_imp = AnnData(X = gt_impute.T, obs = sample_info.loc[sample])
```

```python
# every other low dimensional visualization is based on PCA; pca is already slow for 15000 snps
sc.tl.pca(adata)
sc.tl.pca(adata_imp)
```

```python
# use nearest neighbor way to fill -1 entry
sc.pp.neighbors(adata, n_neighbors=10, n_pcs=20)
sc.pp.neighbors(adata_imp, n_neighbors=10, n_pcs=20)
```

# second way of imputing relies on nearest neighbor in adata.uns

```python
from collections import Counter
def impute_nan(adata, nan_value = -1):
    ''' impute nan using most common value in kNN, not iteratively (so faster)'''
    for person_id in range(len(adata.obs.index)):
        need_impute = np.where(adata.X[person_id, :] == -1)[0]
        neighbor_ids = np.where(adata.uns['neighbors']['connectivities'][person_id].toarray()>0)[1]
        
        for snp_id in need_impute:
            # most common
            neighbors = adata.X[neighbor_ids, snp_id]
            #print(neighbors)
            useless_id = np.where(neighbors == -1)
            neighbors = np.delete(neighbors, useless_id)
            if len(neighbors) > 0:
                neighbor_snp = np.argmax(np.bincount(neighbors.astype(int)))
            else:
                neighbor_snp = np.random.randint(3, size=1)[0]
            adata.X[person_id, snp_id] = neighbor_snp
```

```python
impute_nan(adata)
```

# after imputing we need to recalculate PCA

```python
# redo after imputation
sc.tl.pca(adata)
sc.pp.neighbors(adata, n_neighbors=10, n_pcs=20, use_rep = 'X')
```

```python
# nearest neighbor graph is the basis of umap

sc.tl.draw_graph(adata)
sc.pl.draw_graph(adata, color = ['super_pop','pop', 'batch'])
```

```python
#sc.pp.neighbors(adata_imp, n_neighbors=4, n_pcs=20)
sc.tl.draw_graph(adata_imp)
sc.pl.draw_graph(adata_imp, color = ['super_pop','pop', 'batch'])
```

```python

# customize color
adata.uns['super_pop_colors'] = adata.obs['super_pop'].cat.categories.map(super_pop_color)
adata.uns['pop_colors'] = adata.obs['pop'].cat.categories.map(pop_to_color)
```

```python
sc.pl.draw_graph(adata, color = ['pop', 'super_pop'], legend_loc = 'right margin', ncols = 1)
```

```python
# do umap

sc.tl.umap(adata)
sc.pl.umap(adata, color = ['pop', 'super_pop'], legend_loc = 'right margin', ncols = 1)
```

```python
# to tsne
sc.tl.tsne(adata)
sc.pl.tsne(adata, color = ['pop', 'super_pop'], legend_loc = 'right margin', ncols = 1)
```

```python
# do PAGA
sc.tl.paga(adata, groups='super_pop')
sc.pl.paga(adata, color=['super_pop'], max_edge_width = 2, node_size_scale= 4)
```

```python
# do PAGA
_, axs = plt.subplots(ncols=1, figsize=(15, 5))
sc.tl.paga(adata, groups='pop')
sc.pl.paga(adata, color=['pop'], max_edge_width = 1, node_size_scale= 1, layout = 'fr', ax = axs, fontoutline = 1)
```
